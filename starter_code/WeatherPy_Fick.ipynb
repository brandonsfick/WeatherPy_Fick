```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm
import pprint
import json
from pprint import pprint
# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities

lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=2500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# del cities[160:]
# Print the city count to confirm sufficient count
len(cities)

```

```python
df_cities= pd.DataFrame({'City Name':cities})

df_cities['Temperature'] = ""
df_cities['Humidity'] = ""
df_cities['Cloudiness'] = ""
df_cities['Wind Speed'] = ""
df_cities.head()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# Creates a setting dictionary with information we're interested in


current_weather_city = []
settings = {"units": "imperial", "appid": api_key}
# Get current weather from randomized city list
record_numb = 1
index = 0
print(f"Beginning Data Retrieval \n"
    "-----------------------------")

df_cities['Temperature'] = ""
df_cities['Humidity'] = ""
df_cities['Cloudiness'] = ""
df_cities['Wind Speed'] = ""
df_cities['Longitude'] = ""
df_cities['Latitude'] = ""

for city in cities:
    try:
        current_weather_city = owm.find_city(city, **settings)
        df_cities.loc[index,'Cloudiness'] = current_weather_city[0]['clouds']['all']
        df_cities.loc[index,'Temperature'] = current_weather_city[0]['main']['temp_max']
        df_cities.loc[index,'Humidity'] = current_weather_city[0]['main']['humidity']
        df_cities.loc[index,'Wind Speed'] = current_weather_city[0]['wind']['speed']
        df_cities.loc[index,'Longitude'] = current_weather_city[0]['coord']['lat']
        df_cities.loc[index,'Latitude'] = current_weather_city[0]['coord']['lon']
        print(f"Procesing Record {record_numb}: {city}")
        index = index +1
        record_numb = record_numb+1
        
    except:     
        index = index +1
        print(f"Skipping data for {city} it doesn't have any data.")
    time.sleep(1.2)
```

```python
df_cities.replace('', np.nan, inplace=True)
df_cities.head(141)
```

```python
df_cities=df_cities.dropna(how='any')
df_cities=df_cities.reset_index(drop=True)
df_cities.head(60)
```
